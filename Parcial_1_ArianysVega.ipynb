Ejemplo 3.  un gas ideal que ocupa un volumen de 2.5 litros a una temperatura de 25¬∞C  y una presi√≥n de 1 atm. Si la masa de este gas es de 3.2 gramos, ¬øcu√°l es su masa molar?

4.	Funci√≥n que calcule las concentraciones del √°cido y base conjugada para la preparaci√≥n de un buffer.
Parametros:
pH deseado
pKa constante de acidez
volumen en mL de la soluci√≥n que se prepara
√°cido monopr√≥tico.
**Ecuaci√≥n de Henderson-Hasselbalch:**
pH = pKa + log ([A-]/[HA])
[A-]/[HA]= 10^pH-pKa



Ejemplo 4. Preparaci√≥n de un Buffer de √Åcido Ac√©tico/Acetato
pH deseado = 4.75
pKa del √°cido ac√©tico = 4.76
Volumen total de la soluci√≥n = 500 mL
La suma de las concentraciones de √°cido y base conjugada en el buffer es de 0.1 M

```python
pH=4.75
pKa=4.76
relacion_base_acido = 10 ** (pH - pKa)
print(relacion_base_acido)
concentracion_total = 0.1
# [Base Conjugada] = relacion_base_acido * [√Åcido]
# [√Åcido] + relacion_base_acido * [√Åcido] = 0.1 M
# [√Åcido] * (1 + relacion_base_acido) = 0.1 M
# [√Åcido] = 0.1 / (1 + relacion_base_acido)
# [Base Conjugada] = relacion_base_acido * [√Åcido]
concentracion_acido = concentracion_total / (1 + relacion_base_acido)
print(concentracion_acido)
concentracion_base_conjugada = relacion_base_acido * concentracion_acido
print(concentracion_base_conjugada)



```

8. Calcule la presi√≥n osm√≥tica a 25¬∞C
Parametros:
Concentraci√≥n en g/L
Peso molecular del soluto
R constante
Temperatura
Ecuacion:
ùúã= C.R.T/M
C= concentracion (g/L)
R= constante de los gases (atm¬∑L/mol¬∑K)
T= temperatura en k
M= peso molecular del soluto (g/mol)

Ejemplo 8. Un estudiante de qu√≠mica quiere determinar la presi√≥n osm√≥tica de una soluci√≥n a 25¬∞C. La soluci√≥n contiene un soluto con una concentraci√≥n de 5 g/L. El peso molecular del soluto es 180 g/mol. Utilizando la constante de los gases R= 0.0821 ( atm¬∑L/mol¬∑K). Calcula la presi√≥n osm√≥tica de la soluci√≥n.

```python
temperatura_C=25
temperatura_K = temperatura_C + 273.15
print(temperatura_K)
concentracion = 5
peso_molecular = 180
R=0.0821
presion_osmotica = (concentracion * R * temperatura_K) / peso_molecular
print(presion_osmotica)
```
