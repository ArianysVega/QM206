4.	FunciÃ³n que calcule las concentraciones del Ã¡cido y base conjugada para la preparaciÃ³n de un buffer.
Parametros:
pH deseado
pKa constante de acidez
volumen en mL de la soluciÃ³n que se prepara
Ã¡cido monoprÃ³tico.
**EcuaciÃ³n de Henderson-Hasselbalch:**
pH = pKa + log ([A-]/[HA])
[A-]/[HA]= 10^pH-pKa



Ejemplo 4. PreparaciÃ³n de un Buffer de Ãcido AcÃ©tico/Acetato
pH deseado = 4.75
pKa del Ã¡cido acÃ©tico = 4.76
Volumen total de la soluciÃ³n = 500 mL
La suma de las concentraciones de Ã¡cido y base conjugada en el buffer es de 0.1 M

```python
pH=4.75
pKa=4.76
relacion_base_acido = 10 ** (pH - pKa)
print(relacion_base_acido)
concentracion_total = 0.1
# [Base Conjugada] = relacion_base_acido * [Ãcido]
# [Ãcido] + relacion_base_acido * [Ãcido] = 0.1 M
# [Ãcido] * (1 + relacion_base_acido) = 0.1 M
# [Ãcido] = 0.1 / (1 + relacion_base_acido)
# [Base Conjugada] = relacion_base_acido * [Ãcido]
concentracion_acido = concentracion_total / (1 + relacion_base_acido)
print(concentracion_acido)
concentracion_base_conjugada = relacion_base_acido * concentracion_acido
print(concentracion_base_conjugada)



```

8. Calcule la presiÃ³n osmÃ³tica a 25Â°C
Parametros:
ConcentraciÃ³n en g/L
Peso molecular del soluto
R constante
Temperatura
Ecuacion:
ðœ‹= C.R.T/M
C= concentracion (g/L)
R= constante de los gases (atmÂ·L/molÂ·K)
T= temperatura en k
M= peso molecular del soluto (g/mol)

Ejemplo 8. Un estudiante de quÃ­mica quiere determinar la presiÃ³n osmÃ³tica de una soluciÃ³n a 25Â°C. La soluciÃ³n contiene un soluto con una concentraciÃ³n de 5 g/L. El peso molecular del soluto es 180 g/mol. Utilizando la constante de los gases R= 0.0821 ( atmÂ·L/molÂ·K). Calcula la presiÃ³n osmÃ³tica de la soluciÃ³n.

```python
temperatura_C=25
temperatura_K = temperatura_C + 273.15
print(temperatura_K)
concentracion = 5
peso_molecular = 180
R=0.0821
presion_osmotica = (concentracion * R * temperatura_K) / peso_molecular
print(presion_osmotica)
```
